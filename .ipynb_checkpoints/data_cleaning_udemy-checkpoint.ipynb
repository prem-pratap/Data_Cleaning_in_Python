```python
df1=pd.read_csv("/home/prem/Desktop/courses/data_cleaning.csv")
df1.columns=df1.columns.str.strip().str.lower().str.replace(' ','_')
df1.head()
```

# Outliers

```python
#outliers are the values which are too high not within the range
import seaborn as sb
outliers_df=pd.read_csv('http://bit.ly/CountryDataset1')
outliers_df.head()
```

```python
outlier_df_2007=outliers_df[outliers_df['year']==2007]
outlier_df_2007.head()
```

```python
#palette is the color palette here we chose colorblind friendly palette

boxplot=sb.boxplot(y="lifeExp",x="continent",data=outlier_df_2007,width=1,palette='colorblind')
boxplot.set_title("life expectancy in the world",fontsize=1)
boxplot.set_xlabel("Continents",fontsize=1)
boxplot.set_ylabel("LifeExp",fontsize=1)
```

```python
'''
to deal with the outliers we can set interquartile range so middle 50s will be visible
'''
q1=outliers_df.quantile(0.25)
q3=outliers_df.quantile(0.75)
IQR=q3-q1
print(IQR)
#outlier_iqr=outliers_df[~((outliers_df < (q1-1.5*IQR)) |(outliers_df > (q3+1.5*IQR))).any(axis=1)]
#outlier_iqr.shape
```
